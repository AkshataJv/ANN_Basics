```python
from typing_extensions import ForwardRef
import torch.nn as nn

class ANN(nn.Module):
  def __init__(self):
    super(ANN, self).__init__()

    self.model = nn.Sequential(
        nn.Linear(X.shape[1], 64),
        nn.ReLU(),
        nn.Linear(64, 64),
        nn.ReLU(),
        nn.Linear(64, 7)
    )
  def forward(self, x):
    return self.model(x)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Placeholder for input data X
X = torch.randn(100, 10) # Example: 100 samples, 10 features

model = ANN()

# loss & optim
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
```

```python
# Training ANN

if 'train_loader' not in locals() and 'train_loader' not in globals():

  dummy_y = torch.randint(0, 7, (X.shape[0],))
  dummy_dataset = torch.utils.data.TensorDataset(X, dummy_y)
  train_loader = torch.utils.data.DataLoader(dummy_dataset, batch_size=32, shuffle=True)

epochs = 100
for epoch in range(epochs):
  model.train()

  running_loss = 0.0

  for xb, yb in train_loader:
    optimizer.zero_grad()

    outputs = model(xb)
    loss = criterion(outputs, yb)
    loss.backward()
    optimizer.step()
    # optimizer.zero_grad() # Removed redundant zero_grad

    running_loss += loss.item()

  train_loss = running_loss / len(train_loader)

  print(f"epoch {epoch+1}/{epochs}, Loss: {train_loss:.4f}")
```

```python
# Evaluate
model.eval()

total = 0
correct = 0

with torch.no_grad():
  for xb, yb in train_loader:
    outputs = model(xb)
    _, predicted = torch.max(outputs.data, 1)
    total += yb.size(0)
    correct += (predicted == yb).sum().item()

print("accuracy: ", correct/total*100)
```
